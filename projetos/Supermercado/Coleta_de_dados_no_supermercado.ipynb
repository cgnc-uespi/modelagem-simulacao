Bibliotecas

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import random
import numpy as np
```

A base de dados do supermercado consiste em intevalos sucessivos de chegadas de clientes em minutos.

Criar dataframe a partir de um arquivo csv

```python
# Obtendo os dados
data = pd.read_csv('Coleta de dados do supermercado.csv', header=None)
#data = pd.read_csv('Coleta de dados do supermercado.csv')
#print(data)

data.drop(columns=[1], inplace=True)
#print(data)

df = pd.DataFrame(data)
df
```

```python
max_value = df.max().max()  # Encontra o maior valor da tabela
max_value
```

```python
min_value = df.min().min()  # Encontra o menor valor da tabela
min_value
```

```python
mean = df.mean()
mean
```

```python
median = df.median()
median
```

```python
df_sem_discrepancia = df[~df.isin([max_value]).any(axis=1)]  # Remove a linha que contém esse valor
df_sem_discrepancia
```

Plotar valores coletados

```python
sns.scatterplot(data=df)
plt.xlabel('Intervalos')
plt.ylabel('Minutos')
plt.title('Gráfico de dispersão: Intervalos vs Valores')
plt.show()
```

Valores sem discrepância

```python
sns.scatterplot(data=df_sem_discrepancia)
plt.xlabel('Intervalos')
plt.ylabel('Minutos')
plt.title('Gráfico de dispersão: Intervalos vs Valores')
plt.show()
```

Histograma

```python
class_count = 10

sns.histplot(data=df,bins=class_count,color='blue',edgecolor='black')
plt.xlabel('Intervalos')
plt.ylabel('Frequência')
plt.title('Histograma do supermercado')
plt.show()
```

Sem discrepância

```python
class_count = 10

sns.histplot(data=df_sem_discrepancia,bins=class_count,color='blue',edgecolor='black')
plt.xlabel('Intervalos')
plt.ylabel('Frequência')
plt.title('Histograma do supermercado')
plt.show()
```

Obter a frequência de cada classe no histograma

```python
frequencies, _ = np.histogram(df,bins=class_count)
frequencies
```

Sem discrepância

```python
frequencies, _ = np.histogram(df_sem_discrepancia,bins=class_count)
frequencies
```

Curva da média

```python
mean = pd.DataFrame(columns=['Média'])
for i in range(len(df)):
  mean.loc[i,'Média'] = df.iloc[:i+1, 0].mean()
mean
```

Sem discrepância

```python
mean = pd.DataFrame(columns=['Média'])
for i in range(len(df_sem_discrepancia)):
  mean.loc[i,'Média'] = df_sem_discrepancia.iloc[:i+1, 0].mean()
mean
```

Encontrar a mediana

```python
median = df.median()
median
```

Sem discrepância

```python
median = df_sem_discrepancia.median()
median
```

Plotar a curva

```python
plt.axhline(y=median[0],color='r',linestyle='--',label='Mediana')
sns.lineplot(data=mean)
plt.xlabel('Intervalos')
plt.ylabel('Média')
plt.title('Médias por intervalos')
plt.show()
```
